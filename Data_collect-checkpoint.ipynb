```python
import tushare as ts
import os
import pandas as pd
```

```python
today=ts.get_today_all()
```

```python
oldlist=os.listdir("E://notebook/shsz_stockdata/alldata")
oldlist=pd.DataFrame(oldlist)
oldlist.columns=["code"]
```

```python
def simplecode(x):
    return x[2:8]
```

```python
oldlist=oldlist["code"].apply(simplecode)
oldlist=oldlist.to_list()
newlist=today["code"].to_list()
alllist=[]
alllist.extend(oldlist)
alllist.extend(newlist)
alllistset=set(alllist)
alllist=list(alllistset)
```

```python
print(len(alllist))
```

```python
newdir="E://notebook/shsz_stockdata/data20190617"
i=0
errorlist=[]
for eachcode in alllist:
    try:
        data=ts.get_hist_data(eachcode)
        datapath=os.path.join(newdir,eachcode)
        data.to_csv(datapath)
        i+=1
        if i%20==0:
            print(i)
    except:
        errorlist.extend(eachcode)
        continue
print(errorlist)
print(len(errorlist))
```
